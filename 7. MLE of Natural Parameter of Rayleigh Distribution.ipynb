<b> <h1>Maximum Likelihood Estimation (MLE) of Natural Parameter of Rayleigh Distribution </h1></b>
<br><br>
Let's assume that we have done sampling and drawn a sample of N observation from a population which is Rayleigh Distribution and all N observations are <b>Independent and Identically Distributed (IID)</b>.
<br><br>

\begin{equation}
 L() = P(X = x_{1} \cap X = x_{2} \cap X = x_{3} \cap ... \cap X = x_{N}) 
\end{equation}
<br><br>
\begin{equation}
= P(X = x_{1}) . P(X = x_{2}) . P(X = x_{3}) . ... . P(X = x_{N}) 
\end{equation}
<br><br>
\begin{equation}
= \frac{x_{1}}{\sigma^{\wedge^2}} \mathrm{e}^{-\frac{x_1^2}{2\sigma^{\wedge^2}}} . \frac{x_{2}}{\sigma^{\wedge^2}} \mathrm{e}^{-\frac{x_2^2}{2\sigma^{\wedge^2}}} . ... . \frac{x_{N}}{\sigma^{\wedge^2}} \mathrm{e}^{-\frac{x_N^2}{2\sigma^{\wedge^2}}}
\end{equation}
<br><br>
\begin{equation}
= (\frac{1}{\sigma^{\wedge^2}})^N . \prod_{i=1}^n x_{i}\mathrm{e}^{-\frac{\sum_{i=1}^{N} x_i^2}{2\sigma^{\wedge^2}}}
\end{equation}
<br><br>
\begin{equation}
= (\prod_{i=1}^n x_{i}).(\frac{1}{\sigma^{\wedge^2}})^N \mathrm{e}^{-\frac{\sum_{i=1}^{N} x_i^2}{2\sigma^{\wedge^2}}}
\end{equation}
<br><br>
\begin{equation}
-log_{e} L(\sigma^\wedge)= - {log(\prod_{i=1}^n x_{i})+ log(\frac{1}{\sigma^{\wedge^2}})^N -\frac{\sum_{i=1}^{N} x_i^2}{2\sigma^{\wedge^2}}}
\end{equation}
<br><br>
\begin{equation}
= - [\sum_{i=1}^{N} log_{e} x_{i}+ log_{e}(\frac{1}{\sigma^{\wedge^2}})^N -\frac{\sum_{i=1}^{N} x_i^2}{2\sigma^{\wedge^2}}]
\end{equation}
<br><br>
\begin{equation}
= - \sum_{i=1}^{N} log_{e} x_{i}+ log_{e}\sigma^{\wedge^2N} + \frac{\sum_{i=1}^{N} x_i^2}{2\sigma^{\wedge^2}}
\end{equation}
<br><br>
\begin{equation}
\frac{\delta}{\delta \sigma^\wedge}[-log_{e} L(\sigma^\wedge)]= - \sum_{i=1}^{N} log_{e} x_{i}+ log_{e}\sigma^{\wedge^2N} + \frac{\sum_{i=1}^{N} x_i^2}{2\sigma^{\wedge^2}}
\end{equation}
<br><br>
\begin{equation}
= - \sum_{i=1}^{N} log_{e} x_{i}+ 2N . log_{e}\sigma^{\wedge} + \frac{\sum_{i=1}^{N} x_i^2}{2\sigma^{\wedge^2}}
\end{equation}
<br><br>
\begin{equation}
= 0 + \frac{2N}{\sigma^{\wedge}} + \frac{\sum_{i=1}^{N} x_i^2}{\sigma^{\wedge^3}}
\end{equation}
<br><br>
\begin{equation}
= \frac{\sum_{i=1}^{N} x_i^2}{\sigma^{\wedge^2}} = 2N 
\end{equation}
<br><br>
\begin{equation}
= \frac{\sum_{i=1}^{N} x_i^2}{2N} = \sigma^{\wedge^2}
\end{equation}
<br><br>
<b> Minimum Variance Unbaised (MVU) Estimator of Natural Parameter of Rayleigh Distribution </b>
<br>
\begin{equation}
\sigma_{ML}^{\wedge} = \sqrt{\frac{\sum_{i=1}^{N} x_i^2}{2N}} 
\end{equation}
<br><br>
